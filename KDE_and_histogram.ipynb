```python
#plot histogram and KDE
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
import pylab as py 
```

```python
n=200
sample = np.random.normal(size=n)
plt.hist(sample,bins=20)


```

```python
#kde library implementation
sns.kdeplot(sample,kernel='gau')
```

```python
#uniform distribution
u_sample = np.random.uniform(size=500)
plt.hist(u_sample)
```

```python
sns.kdeplot(u_sample, kernel='gau')
```
