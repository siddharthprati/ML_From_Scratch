```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
import pylab as py 
```

```python
#Compare QQ plot of uniform and normal distribution
n_points = 40000
uniform_points = np.random.uniform(size=n_points)
normal_points = np.random.normal(size=n_points)
sns.distplot(uniform_points)
sns.distplot(normal_points)

```

```python
#plot qq plot

uniform_points = np.sort(uniform_points)
normal_points = np.sort(normal_points)
X=[]
Y=[]
for i in range(0,100):
    X.append(uniform_points[int(n_points/100)*i])
    Y.append(normal_points[int(n_points/100)*i])

plt.plot(Y,X)
plt.show()
```

```python
#plot qq plot

normal_points2 = np.random.normal(loc=4,scale=2,size=n_points)
normal_points = np.sort(normal_points)
normal_points2 = np.sort(normal_points2)
X=[]
Y=[]
for i in range(0,100):
    X.append(normal_points2[int(n_points/100)*i])
    Y.append(normal_points[int(n_points/100)*i])

plt.plot(Y,X)
plt.show()
```

```python
sm.qqplot(uniform_points) 
py.show() 
```
