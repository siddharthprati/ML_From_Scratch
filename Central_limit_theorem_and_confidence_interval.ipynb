```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
import pylab as py 
import pandas as pd
import random as r
```

```python

points = np.random.uniform(size=2000)
sample_list=[]


n_samples = 1000
mean_list=[]
for i in range(0,n_samples):
    temp=[]
    temp = np.random.choice(points,size=30,replace = False)
    sample_list.append(temp)
    mean_list.append(np.mean(temp))
sns.distplot(mean_list)
```

```python

points = np.random.lognormal(size=2000)
sample_list=[]


n_samples = 1000
mean_list=[]
for i in range(0,n_samples):
    temp=[]
    temp = np.random.choice(points,size=100,replace = False)
    sample_list.append(temp)
    mean_list.append(np.mean(temp))
sns.distplot(mean_list)
```

```python
sm.qqplot(np.sort(mean_list))
py.show()
```

```python
input_points = np.random.uniform(size=2000)
#bootstraping method to calculate CI
n_samples = 100
median_list=[]
for i in range(0,n_samples):
    temp=[]
    temp = np.random.choice(input_points,size=30,replace = True)
    sample_list.append(temp)
    median_list.append(np.median(temp))
median_list=np.sort(median_list)
print('96% connfidence interval of median :',median_list[2],median_list[97])
```
