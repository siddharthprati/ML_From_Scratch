## Power law Distribution and Box Cox

#### This notebook will try to show power law distribution and transformation to gaussian

```python
#Power law distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
import scipy.stats as scp
import pylab as py 
import pandas as pd
import random as r
```

```python
np.random.seed(0)
random_pareto = np.random.pareto(a=1,size=100)
sns.distplot(random_pareto)
```

```python
sm.qqplot(random_pareto)
```

```python
transformed_pareto, lamda = scp.boxcox(random_pareto)
sns.distplot(transformed_pareto)
```

```python
#verify whether it is normal or not
sm.qqplot(transformed_pareto)
```

```python
#experiment with lognormal
log_normal_sample = np.random.lognormal(size=300)
sns.distplot(log_normal_sample)
```

```python
sm.qqplot(log_normal_sample)
```

```python
transformed_lg , lamda2 = scp.boxcox(log_normal_sample)
sm.qqplot(transformed_lg)
```
