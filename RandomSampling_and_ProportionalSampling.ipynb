```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm 
import pylab as py 
import pandas as pd
from sklearn.datasets import load_boston
import random as r
```

```python
b = pd.DataFrame(load_boston().data)
b.columns = load_boston().feature_names
print(b.head(8))
print(b.shape)
```

```python
#random_sampling_of_boston_dataset

random_sample = b.sample(frac=0.2,replace=False,random_state=0)
random_sample
```

```python
#proportional sampling
weight = [0.05,0.05,0.1,0.15,.02,0.03,.5,0.07,0.03]
for i in range(1,498):
    weight.append(0)

prop_sample = b.sample(frac=0.02,replace=True,weights= weight)
prop_sample

```
